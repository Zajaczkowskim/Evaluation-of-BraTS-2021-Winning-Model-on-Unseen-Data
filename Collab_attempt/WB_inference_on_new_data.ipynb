```python
###################################################################
#                         FOR TRAINING                            #
###################################################################

import os
import shutil
import random
import json

from pathlib import Path

# Ścieżki wejściowe i wyjściowe
input_dir = "/ASNR-MICCAI-BraTS2023-GLI-Challenge-TrainingData/ASNR-MICCAI-BraTS2023-GLI-Challenge-TrainingData"
task_name = "Task01_BrainTumour"
base_output = "/content/nnUNet_raw_data_base/nnUNet_raw_data"
task_dir = os.path.join(base_output, task_name)

imagesTr_dir = os.path.join(task_dir, "imagesTr")
imagesTs_dir = os.path.join(task_dir, "imagesTs")
labelsTr_dir = os.path.join(task_dir, "labelsTr")
os.makedirs(imagesTr_dir, exist_ok=True)
os.makedirs(imagesTs_dir, exist_ok=True)
os.makedirs(labelsTr_dir, exist_ok=True)

# Losowanie n folderów
all_subdirs = [d for d in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, d))]
random.seed(42)
selected_subdirs = random.sample(all_subdirs, 30)

# Mapa modalności
modality_map = {
    "t2f": 0,
    "t1n": 1,
    "t1c": 2,
    "t2w": 3
}

# Przetwarzanie folderów
for subdir in selected_subdirs:
    full_path = os.path.join(input_dir, subdir)

    for file in os.listdir(full_path):
        if not file.endswith(".nii.gz"):
            continue

        filepath = os.path.join(full_path, file)

        if "seg" in file:
            # Przetwarzanie pliku segmentacji
            patient_id = subdir.split("-")[-2]  # np. BraTS-GLI-00349-000 => 00349
            new_filename = f"BRATS_{patient_id}.nii.gz"
            shutil.copy(filepath, os.path.join(labelsTr_dir, new_filename))
        else:
            # Przetwarzanie modalności
            modality = file.split("-")[-1].split(".")[0]
            if modality not in modality_map:
                continue

            index = modality_map[modality]
            new_filename = f"{subdir}_{index:04}.nii.gz"
            shutil.copy(filepath, os.path.join(imagesTs_dir, new_filename))

print("✅ Pliki zostały przekonwertowane do imagesTs i labelsTr.")

# Tworzenie dataset.json
dataset_json = {
    "name": "BraTS2023_GLI_Training",
    "description": "Converted BraTS2023 GLI Validation data for nnUNet inference",
    "tensorImageSize": "4D",
    "modality": {
        "0": "FLAIR",
        "1": "T1w",
        "2": "t1gd",
        "3": "T2w"
    },
    "labels": {
    "0": "background",
    "1": "edema",
    "2": "non-enhancing tumor",
    "4": "enhancing tumor"
    },
    "numTest": len(selected_subdirs),
    "test": [f"./imagesTs/{s}_{m:04}.nii.gz" for s in selected_subdirs for m in range(4)],
    "numTraining": 0,
    "training": []
}

with open(os.path.join(task_dir, "dataset.json"), "w") as f:
    json.dump(dataset_json, f, indent=4)

print("✅ dataset.json został utworzony.")
```

```python
from IPython.display import display, Markdown

def create_markdown_report(df):
    report = ["## Segmentation Performance Report",
              "| Class | Dice | Accuracy | FP Rate | FN Rate |",
              "|-------|------|----------|---------|---------|"]

    for class_id, group in df.groupby('Class'):
        dice = group['Dice'].mean()
        acc = group['Accuracy'].mean()
        fp = group['False Positive Rate'].mean()
        fn = group['False Negative Rate'].mean()

        report.append(f"| {class_id} | {dice:.3f} | {acc:.3f} | {fp:.3f} | {fn:.3f} |")

    return "\n".join(report)

display(Markdown(create_markdown_report(df)))
```

```python
import nibabel as nib
import numpy as np

def load_nifti(filepath):
    return nib.load(filepath).get_fdata().astype(np.int32)

def compute_dice_per_class(gt, pred, num_classes):
    dice_scores = {}
    for c in range(num_classes):
        gt_c = (gt == c)
        pred_c = (pred == c)

        intersection = np.logical_and(gt_c, pred_c).sum()
        volume_sum = gt_c.sum() + pred_c.sum()

        if volume_sum == 0:
            dice = 1.0  # Class not present in either mask
        else:
            dice = 2.0 * intersection / volume_sum

        dice_scores[c] = dice
    return dice_scores

# File paths
gt_path = "/content/nnUNet_raw_data_base/nnUNet_raw_data/Task02_BrainSSA/labelsTr/BraTS-SSA-00095-000.nii.gz"
pred_path = "/content/OUTPUT_DIRECTORY/BraTS-SSA-00095-000.nii.gz"

# Load NIfTI masks
gt_mask = load_nifti(gt_path)
pred_mask = load_nifti(pred_path)

# Number of classes (e.g., 0 to N-1); change this to fit your data
num_classes = int(max(gt_mask.max(), pred_mask.max()) + 1)

# Compute Dice for each class
dice_scores = compute_dice_per_class(gt_mask, pred_mask, num_classes)

# Print results
for cls, score in dice_scores.items():
    print(f"Class {cls}: Dice Score = {score:.4f}")
```

```python
import os
import nibabel as nib
import numpy as np
from sklearn.metrics import f1_score

def load_nifti(path):
    """Load NIFTI file and return data as numpy array"""
    return nib.load(path).get_fdata()

def compute_metrics(pred_dir, label_dir):
    metrics = []

    # Get matching filenames (assuming same names in both dirs)
    pred_files = sorted(os.listdir(pred_dir))
    label_files = sorted(os.listdir(label_dir))

    for pred_file, label_file in zip(pred_files, label_files):
        # Load data
        pred = load_nifti(os.path.join(pred_dir, pred_file))
        label = load_nifti(os.path.join(label_dir, label_file))

        # Create BraTS regions
        pred_et = (pred == 3).astype(np.uint8)
        pred_tc = np.isin(pred, [2, 3]).astype(np.uint8)
        pred_wt = np.isin(pred, [1, 2, 3]).astype(np.uint8)

        label_et = (label == 3).astype(np.uint8)
        label_tc = np.isin(label, [2, 3]).astype(np.uint8)
        label_wt = np.isin(label, [1, 2, 3]).astype(np.uint8)

        # Compute Dice scores
        metrics.append({
            'case': pred_file,
            'ET': f1_score(label_et.flatten(), pred_et.flatten()),
            'TC': f1_score(label_tc.flatten(), pred_tc.flatten()),
            'WT': f1_score(label_wt.flatten(), pred_wt.flatten())
        })

    return metrics

# Usage
metrics = compute_metrics("/content/Output_data", "/content/labelsTr")

# Print results
import pandas as pd
df = pd.DataFrame(metrics)
print("Average Dice Scores:")
print(df[['ET', 'TC', 'WT']].mean())
```

```python
def compute_brats_metrics(pred, label):
    """Compute all BraTS metrics for a single case"""
    results = {}

    # Define regions
    regions = {
        'ET': ([3], [3]),
        'TC': ([2,3], [2,3]),
        'WT': ([1,2,3], [1,2,3])
    }

    for name, (pred_classes, label_classes) in regions.items():
        y_pred = np.isin(pred, pred_classes).astype(np.uint8)
        y_true = np.isin(label, label_classes).astype(np.uint8)

        eps = 1e-7  # prevent division by zero
        tp = np.sum(y_true * y_pred)
        fp = np.sum(y_pred) - tp
        fn = np.sum(y_true) - tp

        results[name] = {
            'Dice': (2*tp + eps) / (2*tp + fp + fn + eps),
            'Sensitivity': tp / (tp + fn + eps),
            'Specificity': np.sum((1-y_true)*(1-y_pred)) / (np.sum(1-y_true) + eps),
            'Volume_diff': np.abs(np.sum(y_pred) - np.sum(y_true))
        }

    return results
```

```python
import os
import nibabel as nib
import numpy as np
from tqdm import tqdm  # for progress bar

def compute_brats_metrics_avg(pred_dir, label_dir):
    """Compute average BraTS metrics across all files in directories"""
    # Initialize metrics accumulators
    metrics_sum = {
        'ET': {'Dice': 0, 'Sensitivity': 0, 'Specificity': 0, 'Volume_diff': 0},
        'TC': {'Dice': 0, 'Sensitivity': 0, 'Specificity': 0, 'Volume_diff': 0},
        'WT': {'Dice': 0, 'Sensitivity': 0, 'Specificity': 0, 'Volume_diff': 0}
    }
    num_cases = 0

    # Get matching files (assuming matching filenames)
    pred_files = sorted([f for f in os.listdir(pred_dir) if f.endswith('.nii.gz')])
    label_files = sorted([f for f in os.listdir(label_dir) if f.endswith('.nii.gz')])

    for pred_file, label_file in tqdm(zip(pred_files, label_files), total=len(pred_files)):
        # Load data
        pred = nib.load(os.path.join(pred_dir, pred_file)).get_fdata()
        label = nib.load(os.path.join(label_dir, label_file)).get_fdata()

        # Define regions
        regions = {
            'ET': ([3], [3]),
            'TC': ([2,3], [2,3]),
            'WT': ([1,2,3], [1,2,3])
        }

        # Compute metrics for each region
        for name, (pred_classes, label_classes) in regions.items():
            y_pred = np.isin(pred, pred_classes).astype(np.uint8)
            y_true = np.isin(label, label_classes).astype(np.uint8)

            eps = 1e-7  # prevent division by zero
            tp = np.sum(y_true * y_pred)
            fp = np.sum(y_pred) - tp
            fn = np.sum(y_true) - tp
            tn = np.sum((1-y_true)*(1-y_pred))

            # Accumulate metrics
            metrics_sum[name]['Dice'] += (2*tp + eps) / (2*tp + fp + fn + eps)
            metrics_sum[name]['Sensitivity'] += tp / (tp + fn + eps)
            metrics_sum[name]['Specificity'] += tn / (tn + fp + eps)
            metrics_sum[name]['Volume_diff'] += np.abs(np.sum(y_pred) - np.sum(y_true))

        num_cases += 1

    # Compute averages
    avg_metrics = {}
    for region in metrics_sum:
        avg_metrics[region] = {
            k: v / num_cases
            for k, v in metrics_sum[region].items()
        }

    return avg_metrics

# Usage example
avg_metrics = compute_brats_metrics_avg(
    pred_dir="/content/Output_data/content/OUTPUT_DIRECTORY",
    label_dir="/content/nnUNet_raw_data_base/nnUNet_raw_data/Task01_BrainTumour/labelsTr"
)

# Pretty print results
import pandas as pd
print(pd.DataFrame(avg_metrics).T.round(3))
```

```python
import os
import nibabel as nib
import numpy as np
import pandas as pd
from collections import defaultdict

def validate_labels(pred_dir, expected_labels=[0, 1, 2, 3]):
    """
    Sprawdza czy predykcje zawierają tylko oczekiwane etykiety

    Args:
        pred_dir (str): Ścieżka do folderu z plikami .nii.gz
        expected_labels (list): Lista dopuszczalnych etykiet (np. [0,1,2,3])

    Returns:
        tuple: (DataFrame ze statystykami, listę plików z nieprawidłowymi etykietami)
    """
    invalid_files = []
    stats = []

    for filename in os.listdir(pred_dir):
        if filename.endswith('.nii.gz'):
            filepath = os.path.join(pred_dir, filename)
            pred = nib.load(filepath).get_fdata()
            unique_labels = np.unique(pred).astype(int)

            # Sprawdź nieoczekiwane etykiety
            unexpected_labels = set(unique_labels) - set(expected_labels)
            has_unexpected = len(unexpected_labels) > 0

            if has_unexpected:
                invalid_files.append({
                    'filename': filename,
                    'unexpected_labels': list(unexpected_labels),
                    'unexpected_counts': {label: np.sum(pred == label) for label in unexpected_labels}
                })

            # Zbierz statystyki
            stats.append({
                'filename': filename,
                'all_labels': list(unique_labels),
                'has_unexpected': has_unexpected,
                **{f'label_{label}_count': np.sum(pred == label) for label in unique_labels}
            })

    # Konwertuj do DataFrame
    stats_df = pd.DataFrame(stats)

    # Podsumowanie
    print(f"\nZnaleziono {len(invalid_files)} plików z nieoczekiwanymi etykietami")
    if len(invalid_files) > 0:
        print("\nPliki z problemami:")
        for file in invalid_files:
            print(f"{file['filename']}: Nieoczekiwane etykiety {file['unexpected_labels']}")
            for label, count in file['unexpected_counts'].items():
                print(f"   - Etykieta {label}: {count} voxeli ({count/np.prod(pred.shape)*100:.2f}%)")

    # Statystyki globalne
    all_labels = np.unique(np.concatenate(stats_df['all_labels'].values))
    print("\nGlobalna dystrybucja etykiet:")
    for label in all_labels:
        count = stats_df[f'label_{label}_count'].sum()
        print(f"Etykieta {label}: {count} voxeli")

    return stats_df, invalid_files

# Użycie:
pred_dir = "/content/OUTPUT_DIRECTORY"
expected_labels = [0, 1, 2, 3]  # Tylko te etykiety są akceptowalne
stats_df, problem_files = validate_labels(pred_dir, expected_labels)

# Dodatkowa wizualizacja
import matplotlib.pyplot as plt

# Zlicz wystąpienia każdej etykiety
label_counts = {label: stats_df[f'label_{label}_count'].sum()
               for label in expected_labels if f'label_{label}_count' in stats_df}

plt.figure(figsize=(10,5))
plt.bar(label_counts.keys(), label_counts.values())
plt.title("Dystrybucja etykiet w predykcjach")
plt.xlabel("Etykieta")
plt.ylabel("Liczba voxeli")
plt.xticks(list(label_counts.keys()))
plt.show()
```
